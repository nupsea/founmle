```python
# Import dataset

import pandas as pd
df = pd.read_csv("../data/winequality-red.csv", sep=';')
df.head()
```

```python
# Rename columns

df.columns = [col.replace(' ', '_') for col in df.columns]
df.head()
```

```python
# Analyze features

```

```python
# Categorize

def numeric_to_buckets(df, column_name):
    median = df[column_name].median()
    for i, val in enumerate(df[column_name]):
        if val >= median:
            df.loc[i, column_name + '_rating'] = 'high'
        else:
            df.loc[i, column_name + '_rating'] = 'low' 
```

```python
for feature in df.columns[:-1]: # All columns(features) except the quality metric.
    numeric_to_buckets(df, feature)
    print(df.groupby(feature + '_rating').quality.mean(), '\n')
```
